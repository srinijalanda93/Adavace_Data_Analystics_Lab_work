```python
SEARCH_QUERY = "python programming"  # You can change this to any topic
MAX_RESULTS = 12  # Number of videos to fetch (YouTube limits to 50 max per call)
```

```python

youtube = build('youtube', 'v3', developerKey=API_KEY)


request = youtube.search().list(
    q=SEARCH_QUERY,
    part="snippet",
    type="video",
    maxResults=MAX_RESULTS
)

response = request.execute()

video_data = []

for item in response['items']:
    video_info = {
        "title": item['snippet']['title'],
        "videoId": item['id']['videoId'],
        "channel": item['snippet']['channelTitle'],
        "publishedAt": item['snippet']['publishedAt']
    }
    video_data.append(video_info)

print(f"\nTop {len(video_data)} YouTube Results for Query: '{SEARCH_QUERY}'\n")
for video in video_data:
    print(f"Title: {video['title']}")
    print(f"Video ID: {video['videoId']}")
    print(f"Channel: {video['channel']}")
    print(f"Published At: {video['publishedAt']}")
    print("------")
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# step 1: Initialize graph
G = nx.Graph()

# step 2: Add nodes for each unique channel
channels = set(video["channel"] for video in video_data)
G.add_nodes_from(channels)

# step 3: Add edges between all channels that co-occurred in the same search
channel_list = list(channels)
for i in range(len(channel_list)):
    for j in range(i + 1, len(channel_list)):
        G.add_edge(channel_list[i], channel_list[j])


from collections import Counter
channel_counts = Counter(video["channel"] for video in video_data)
nx.set_node_attributes(G, channel_counts, "video_count")

# step 4: Visualize the graph
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G, seed=42)

# node size based on how many times the channel appeared
sizes = [500 + channel_counts.get(node, 1) * 300 for node in G.nodes()]

nx.draw_networkx(
    G, pos,
    with_labels=True,
    node_size=sizes,
    node_color='skyblue',
    edge_color='gray',
    font_size=9
)

plt.title("YouTube Channel Social Graph (based on search: python programming)", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
from collections import Counter


#step 1: Create a counter for channel appearances
channel_counts = Counter(video["channel"] for video in video_data)


G = nx.Graph()

#step 3: Add nodes and attributes
for channel, count in channel_counts.items():
    G.add_node(channel, video_count=count)

#step 4: Add edges between co-occurring channels (fully connected)
channels = list(channel_counts.keys())
for i in range(len(channels)):
    for j in range(i + 1, len(channels)):
        G.add_edge(channels[i], channels[j])

#step 5: Visualization
plt.figure(figsize=(12, 9))
pos = nx.spring_layout(G, seed=42)

# Dynamic node size based on number of videos per channel
sizes = [300 + 500 * channel_counts[node] for node in G.nodes()]

# Node color (optional) by frequency
colors = [channel_counts[node] for node in G.nodes()]

nx.draw_networkx(
    G, pos,
    with_labels=True,
    node_size=sizes,
    node_color=colors,
    cmap=plt.cm.coolwarm,
    font_size=10,
    edge_color='gray'
)

plt.title("YouTube Social Graph: Channel Frequency in 'Python Programming' Search", fontsize=15)
plt.axis('off')
plt.tight_layout()
plt.show()
```
